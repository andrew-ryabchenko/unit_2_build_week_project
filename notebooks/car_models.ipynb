```python
def clean(cars):
    cars = cars
    for key in cars.keys():
        s = cars[key].lower()
        s = s.split('\n')
        s = [string.replace(' ','') for string in s]
        s = list(set(s))
        cars[key] = s
        
    return cars   
```
